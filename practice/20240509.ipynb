```python
import pandas as pd
train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/MedicalCost/train.csv")
train.head()
```

```python
#흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?
smoke = train[train.smoker == 'yes'].charges.quantile(0.9)
nosmoke = train[train.smoker == 'no'].charges.quantile(0.9)

mean_yes = train.loc[(train.smoker == 'yes') & (train.charges >= smoke)].charges.mean()
mean_no = train.loc[(train.smoker == 'no') & (train.charges >= nosmoke)].charges.mean()

answer = mean_yes - mean_no
answer
```
