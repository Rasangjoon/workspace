```python
#103. 매년 5월달의 open가격의 평균값을 데이터 프레임으로 표현하라

df['Date'] = pd.to_datetime(df['Date'])

target = df.groupby(df['Date'].dt.strftime('%Y-%m')).mean()

answer = target.loc[target.index.str.contains('-05')].Open.to_frame()


answer
```

```python
#104. Tm 컬럼은 각 팀의 이름을 의미한다. TOR팀의 평균나이를 소수 4째 자리까지 구하여라

answer = round(df[df.Tm == 'TOR'].Age.mean(),4)

answer
```

```python
#107. PTS컬럼은 경기당 평균득점수 이다. 각포지션별로 경기당 평균득점수의 평균을 구하여라
df['PTS']
```

```python
result  = df.groupby('Pos')['PTS'].mean().sort_values()

result
```

```python
#109. G컬럼은 참여한 경기의 숫자이다. 각 팀별로 가장 높은 경기참여 수를 가진 선수들의 경기 참여 숫자의 평균을 구하여라

result = df.sort_values(['Tm','G']).groupby('Tm').tail(1).G.mean()

result
```

```python
#110. Tm의 값이 MIA이며 Pos는 C또는 PF인 선수의 MP값의 평균은?

result = df[(df.Tm == 'MIA') & (df.Pos.isin(['C','PF']))].MP.mean()

result
```

```python
#111. 전체 데이터중 G의 평균값의 1.5배 이상인 데이터들만 추출했을때 3P값의 평균은?

result = df[df.G >= df.G.mean()*1.5]['3P'].mean()

result
```

```python
#112. Age의 평균 이상인 그룹과 평균 미만인 그룹간의 G값의 평균의 차이는?

result = df[df.Age >= df.Age.mean()].G.mean() - df[df.Age < df.Age.mean()].G.mean()

result

#result = df[df.Age >= df.Age.mean()].G.mean() - df[df.Age < df.Age.mean()].G.mean()
#print(result)
```
